````python
numEpochs = 250
test_loss = np.zeros(numEpochs)

class TestCallback(Callback):
    def __init__(self, test_data):
        self.test_data = test_data

    def on_epoch_end(self, epoch, logs={}):
        x, y = self.test_data
        loss = self.model.evaluate(x, y, verbose=0)
        test_loss[epoch-1] = loss
        print('\nTesting loss: {}'.format(loss))
````

## 2. Deep Networks
Las *deep network*, o lo que hoy en día se conoce como *deep learning*, hace referencia a modelos de redes neuronales estructurados con muchas capas, es decir, el cómputo de la función final es la composición una gran cantidad de funciones ( $f^{(n)} = f^{(n-1)} \circ f^{(n-2)} \circ \cdots \circ f^{(2)} \circ f^{(1)} $ con $n \gg 0$ ).  
Este tipo de redes neuronales tienen una gran cantidad de parámetros, creciendo exponencialmente por capa con las redes *feed forward*, siendo bastante dificiles de entrenar comparadas con una red poco profunda, esto es debido a que requieren una gran cantidad de datos para ajustar correctamente todos esos parámetros. Pero entonces ¿Cuál es el beneficio que tienen este tipo de redes? ¿Qué ganancias trae el añadir capas a una arquitectura de una red neuronal?

La inicialización de los pesos de las redes neuronales es un campo de estudio completo, ya que la inicialización cuidadosa de la red puede acelerar el proceso de aprendizaje. Las bibliotecas modernas de aprendizaje profundo, como Keras, ofrecen una gran cantidad de métodos de inicialización de red, donde la inicialización de los pesos con pequeños números aleatorios. Para cualquier función de activación, el gradiente se hace mas pequeño a medida que se realiza el backpropagation. Las capas anteriores se hacen mas lentas para entrenar. Más específicamente, en el caso de sigmoid, si sus pesos son grandes, entonces el gradiente será muy pequeño, evitando así que los pesos cambien su valor. Se puede presentar el problema del gradiente desvaneciente, donde se hace que sea muy difícil aprender y ajustar los parámetros de las capas anteriores en la red. En este caso con solo 3 capas este problema no es tan evidente, pero este problema empeora a medida que aumenta el número de capas en la arquitectura.
La actualización del peso es menor y resulta en una convergencia más lenta. Esto hace que la optimización de la función de pérdida sea lenta. En el peor de los casos, esto puede detener completamente la formación de la red neuronal.

En problema del punto anterior se vuelve mucho peor en este caso, porque la primera capa mapea una región de entrada grande a una región de salida más bajas, después se mapeara a un región más baja en la segunda capa, y así sucesivamente dando como resultado, que grandes cambios en la entrada no permiten grandes cambios en la salida. Todo esto se debe a que por su arquitectura la función de activación sigmoid tiene la propiedad de “squash” (aplastar o compactar) el espacio de entrada. Esto es un problema para la función de activación sigmoidal, ya que es aproximadamente lineal cerca de 0, lo que significa que no tiene ningún beneficio tener capas múltiples. Por otro lado, sus activaciones se vuelven cada vez más grandes, entonces sus activaciones se saturaran y deja de tener sentido con gradientes que se acercan a 0. 

En este caso, el gradiente desvaneciente se puede observar antes del entrenamiento. Se puede observar cómo los gradientes de la primera capa son mucho más bajos que los de la tercera capa, lo que significa que esos pesos están cambiando en una cantidad mucho menor. Si agregamos más capas, la diferencia solo se vuelve más dramática. El resto de la red se ve afectado por lo que sale de la primera capa, por lo que si esos primeros pesos son totalmente erróneos, nuestra red no tendrá un buen desempeño.

Las dos funciones de activación son muy similares, excepto que Softplus es diferenciable es cero. Por otra parte, RElu hace más fáciles los cálculos y su derivada, por lo que los algoritmos forward pass y backward pass son más rápidos. Con la función de activación Relu se favorecen las representaciones distribuidas, por su arquitectura no es derivable en cero, la derivada no es continua. Sin embargo, no sucede tan frecuentemente que la suma ponderada sea cero, para esto se puede normalizar la data entre 0 y 1. En casos donde no se tengan estas garantías es mejor usar la versión diferenciable Softplus, que además hace el entrenamiento es más sencillo.
